```python
import statsmodels.api as sm
import numpy as np
```

```python
residuals = np.random.randn(2000)
```

```python
import statsmodels
```

```python
statsmodels.graphics.tsaplots.plot_acf(residuals)
```

```python
statsmodels.graphics.tsaplots.plot_pacf(residuals)
```

```python
import scipy.stats as stats
sm.qqplot(residuals, stats.t, fit=True, line='45')
```
