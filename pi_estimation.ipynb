# Estimation of pi value via montecarlo simulation

```python
import random
import math
```

```python
sample_size = 100000 #The higher the num_samples, the more accurate, but speed impacts
radius = 0.5
x = [random.random()-radius for x in range(sample_size)]
y = [random.random()-radius for x in range(sample_size)]
```

```python
count = 0
c_x = []
c_y = []
for iter_ in range(sample_size):
    dist = math.sqrt (x[iter_]**2 + y[iter_]**2)
    if dist <= radius: 
        count += 1
        c_x.append(x[iter_])
        c_y.append(y[iter_])
```

```python
area_of_circle = count/sample_size
pi = area_of_circle/(radius**2) #circle area = pi * radius^2, backsolving for pi
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
fi = figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(x, y, s=2, color='r')
plt.scatter(c_x, c_y, s=2, color='w')
plt.hlines(0, -0.6, 0.6)
plt.vlines(0, -0.6, 0.6)
plt.title('Estimation of pi via Montecarlo. PI = {}, Num. Samples = {}'.format(pi, sample_size))
```
